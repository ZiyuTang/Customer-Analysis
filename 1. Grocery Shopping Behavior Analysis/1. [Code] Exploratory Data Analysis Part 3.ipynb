```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import Data
df = pd.read_csv("/Users/tangziyu/Desktop/c_3_31.csv",index_col=0)

# Prepare Data
x = df.index.values.tolist()
y1 = df['High income'].values.tolist()
y2 = df['Medium income'].values.tolist()
y3 = df['Low income'].values.tolist()

columns = ['Low income', 'Medium income','High income']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(16,9), dpi= 80)
ax.fill_between(x, y1=y1, y2=0, label=columns[2], alpha=1, color='brown', linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[1], alpha=1, color='orange', linewidth=2)
ax.fill_between(x, y1=y3, y2=0, label=columns[0], alpha=1, color='indianred', linewidth=2)

# Decorations
ax.set_title('Average monthly expenditure on grocery among income groups', fontsize=18)
ax.set(ylim=[0,250])
ax.legend(loc='best', fontsize=12)
plt.xticks(df.index,("Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"))
plt.yticks(np.arange(0, 250,10), fontsize=10)
plt.xlim(1, 12)


# Lighten borders
plt.gca().spines["top"].set_alpha(0.3)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0.3)
plt.gca().spines["left"].set_alpha(.3)
fig.savefig('average monthly expenditure.png', dpi=800)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import Data
df = pd.read_csv("/Users/tangziyu/Desktop/c_3_32.csv",index_col=0)

# Prepare Data
x = df.index.values.tolist()
y1 = df['prive_share_1'].values.tolist()
y2 = df['prive_share_2'].values.tolist()
y3 = df['prive_share_3'].values.tolist()

columns = ['Low income', 'Medium income','High income']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(16,9), dpi= 80)
ax.fill_between(x, y1=y1, y2=0, label=columns[0], alpha=1, color='indianred', linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[1], alpha=1, color='orange', linewidth=2)
ax.fill_between(x, y1=y3, y2=0, label=columns[2], alpha=1, color='brown', linewidth=2)

# Decorations
ax.set_title('The % of private label share in monthly expenditures among income groups', fontsize=18)
ax.set(ylim=[0, 0.15])
ax.legend(loc='best', fontsize=12)
plt.xticks(df.index,("Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"))
plt.yticks(np.arange(0, 0.15,0.01), fontsize=10)
plt.xlim(1, 12)


# Lighten borders
plt.gca().spines["top"].set_alpha(0.3)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0.3)
plt.gca().spines["left"].set_alpha(.3)
fig.savefig(' % of private label share.png', dpi=800)
plt.show()
```
